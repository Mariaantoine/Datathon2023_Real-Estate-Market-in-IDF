```python
df_final = pd.read_csv('final_file.csv')
```

```python
df_final.dropna(inplace=True)
```

```python
code_departement_input = 75 #filtre streamlit
type_local_input = 'Appartement'
df_final = df_final[(df_final.code_departement == code_departement_input) | (df_final.type_local == type_local_input) ]
```

```python
df_final.shape
```

```python
List_col = ['valeur_fonciere','code_commune','nombre_pieces_principales',
            'Total_surface','taux pret','duree pret','taux chômage'] 
```

```python
X, y = df_final[List_col], df_final["prix_total"]# Splitting
# Instantiation
```

```python
X['taux chômage'] = X['taux chômage'].apply(lambda x : float(x.replace(',','.')))
```

```python
train_X, test_X, train_y, test_y = train_test_split(X, y,test_size = 0.30, random_state = 42)
```

```python
xgb_r = xg.XGBRegressor(objective ='reg:linear', n_estimators = 1000, seed = 42,learning_rate=0.01,max_depth=3)# Fitting the model
xgb_r.fit(train_X, train_y)# Predict the model
pred_test = xgb_r.predict(test_X)
pred_train = xgb_r.predict(train_X)
# RMSE Computation
print(metrics.r2_score(train_y,pred_train))
print(metrics.r2_score(test_y,pred_test))
print(metrics.mean_absolute_error(test_y,pred_test))
print(m.sqrt(metrics.mean_squared_error(test_y,pred_test)))
```
